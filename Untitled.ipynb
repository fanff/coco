```python
import numpy as np
import time
```

```python
now = time.time()
d = np.arange(0,8,.11)

def posgen(d,freq = .5, phase = .25,ampl=1,shift=0.3):
    return np.sin(freq*(d*2*np.pi + 2*np.pi*phase))*(ampl/2)+shift

df = pd.DataFrame(d,columns=["time"])


rf= {"freq" :.5, "phase" : 0.5,"ampl":1,"shift":0.2}
lf= {"freq" :.5, "phase" : 0.5,"ampl":1,"shift":-0.2}


hr= {"freq" :.5, "phase" : .2,"ampl":.9,"shift":0.2}
hl= {"freq" :.5, "phase" : .2,"ampl":.9,"shift":0.2}

strt = time.time()
df["rf"] = posgen(df["time"],**rf)
df["lf"] = posgen(df["time"],**lf)
df["hr"] = posgen(df["time"],**hr)
df["hl"] = posgen(df["time"],**hl)
print(time.time()-strt)

df.plot("time",["rf","hl"])
```

```python
def forward(freqfact=1,amplfact = 1.0):
    rf= {"freq" :.4*freqfact, "phase" : 0.5,"ampl":1.0*amplfact,"shift":0.2}
    lf= {"freq" :.4*freqfact, "phase" : 0.5,"ampl":1.0*amplfact,"shift":-0.2}


    hr= {"freq" :.4*freqfact, "phase" : .2,"ampl":.9*amplfact,"shift":0.2}
    hl= {"freq" :.4*freqfact, "phase" : .2,"ampl":.9*amplfact,"shift":0.2}
    
    return {"rf":rf,"lf":lf,"hr":hr,"hl":hl}

```

```python
def makepos(iparms,t):
    return {k:posgen(t,**v) for k,v in iparms.items()}
        
makepos(forward(),time.time())
```
